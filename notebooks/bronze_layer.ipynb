```python
import os

FILES = [
    "customers.csv",
    "orders.csv",
    "order_items.csv",
    "products.csv",
    "stores.csv"
]

INPUT_DIR = "/lakehouse/default/Files/raw_data"
BRONZE_DIR = "/lakehouse/default/Files/bronze"

os.makedirs(BRONZE_DIR, exist_ok=True)

for file in FILES:
    input_path = f"{INPUT_DIR}/{file}"
    bronze_path = f"{BRONZE_DIR}/{file}"

    with open(input_path, "r", encoding="utf-8") as f:
        data = f.read()

    with open(bronze_path, "w", encoding="utf-8") as f:
        f.write(data)

    print(f"{file} → Bronze loaded")

print("Bronze Layer completed")
```
